### Oppgave 1 a)

Sannsynlighetstettheten til en normalfordeling er gitt ved

\begin{equation}
f(x)=\frac{1}{\sqrt{2\pi}}\frac{1}{\sigma}e^{-\frac{1}{2}\frac{(x-\mu)^{2}}{\sigma^{2}}}. 
\label{eq:normalfordeling} \tag{1}
\end{equation}

Med oppgitte verdier for forventningsverdi og varians satt inn, gir dette


\begin{equation}
\varphi=\frac{1}{\sqrt{2\pi}}\frac{1}{\sqrt{at}}e^{-\frac{1}{2}\frac{x^{2}}{at}}.
\label{eq:normfordinnsatt} \tag{2}
\end{equation}

Den oppgitte ligningen som skal løses er diffusjonsligningen med konstant D,

\begin{equation}
\frac{\partial\varphi(x,t)}{\partial t} = D\frac{\partial^{2}\varphi}{\partial x^{2}},
\label{eq:diffusjon} \tag{3}
\end{equation}

hvor

\begin{equation}
D = \frac{(\Delta x)^{2}}{2\Delta t}. 
\label{eq:d} \tag{4}
\end{equation}

I oppgave 1 er det oppgitt at både $\Delta x$ og $\Delta t$ skal settes lik 1. Dette gir dermed at 

\begin{equation}
D=\frac{1}{2}. 
\label{eq:dinnsatt} \tag{5}
\end{equation}

For å se om $\varphi(x,t)$ løser diffusjonsligningen, partiellderiveres $\varphi(x,t)$ mhp. x og t. Dette gir uttrykkene

\begin{equation}
\frac{\partial\varphi(x,t)}{\partial t}= \frac{ae^{\frac{-x^{2}}{2at}}(x^{2}-at)}{2\sqrt{2\pi}(at)^{\frac{5}{2}}} 
\label{eq:partiellphit} \tag{6}
\end{equation}

og

\begin{equation}
\frac{\partial^{2}\varphi}{\partial x^{2}}= \frac{e^{\frac{-x^{2}}{2at}}(x^{2}-at)}{\sqrt{2\pi}(at)^{\frac{5}{2}}}. \label{eq:partiellphix} \tag{7}
\end{equation}

Settes $\eqref{eq:partiellphit}$, $\eqref{eq:partiellphix}$ og $\eqref{eq:dinnsatt}$ inn i $\eqref{eq:diffusjon}$ og forkortes, ser man at med $a = 1$ løser $\varphi(x,t)$ diffusjonsligningen.


### Oppgave 1 b)

```python
import numpy as np
from matplotlib import pyplot as plt
import timeit
import statistics
from scipy.optimize import curve_fit
import time
```

```python
dx = 1                                          #Definerer delta x og delta t til å være dim.løse
dt = 1  

def virrevandrer(M, pR):                        #Funksjonen tar inn antall tidssteg M og sannsynligheten for å gå til høyre pR
    posisjon = np.zeros(M+1)                    #Lager et array med M+1 nuller for å fylle med posisjoner fra tid 0 til 10
    tilfeldigeTall = np.random.uniform(0,1,M)   #Genererer M tilfeldige tall mellom 0 og 1 til for forflytning fra tid 1 til M
    tid = np.arange(0,M+1,dt)                   #Lager en array med tid fra 0 til M+1 med steglengde dt

    for i in range(1,M+1):                      #Kjører en for-løkke som fyller posisjonsarrayen fra tid 1 til M
        if tilfeldigeTall[i-1]<pR:              #Hvis det tilfeldige tallet (sannsynligheten) er mindre enn oppgitt pR, går virrevandreren til høyre
            posisjon[i] = posisjon[i-1]+dx
        else:                                   #Ellers går den til venstre
            posisjon[i] = posisjon[i-1]-dx

    return posisjon, tid, M                     #Returnerer posisjonene, tidsarrayen og antall tidssteg M dersom denne skal brukes senere
```

### Oppgave 1 c)

```python
def plottVirrevandrer(N, pR):                    #Skriver plottingen inn i en funksjon slik at antall steg og ønskede sannsynligheter enkelt kan endres
    plt.figure()
    plt.xlabel('Posisjon (x)', fontsize=15)
    plt.ylabel('Tid (t)', fontsize=15)
    plt.title(f'Posisjon x som funksjon av tid t: {len(pR)} simuleringer', fontsize=20)
    
    for i in range(len(pR)):                     #Bruker lengden av pR-listen og henter ut posisjon og tid fra virrevandrerfunksjonen, og plotter disse
        plt.plot(virrevandrer(N,pR[i-1])[0],virrevandrer(N,pR[i-1])[1], label=f'Simulering med pR {pR[i-1]}')
    plt.legend()
    plt.show()

pR = [0.45,0.50,0.55]
plottVirrevandrer(20,pR)
plottVirrevandrer(500,pR)
```

### Oppgave 1 d)

```python
def flereVandrere(N, M, pR):             #Funksjonen tar inn posisjonen om N virrevandrere på M tidspunkt med sannsynlighet om å gå til høyre pR.
    alleVandrere = np.zeros((N,M+1))     #Lager en matrise med nuller i med N rader(vandrere) og M+1 kolonner(tidspunkt) for å få med posisjonen i tid 0 før vandringen starter.

    for i in range(N):                   #Kjører for løkka for N virrevandrere
        virrer = virrevandrer(M, pR)
        virrePosisjon = virrer[0]
        alleVandrere[i] = virrePosisjon  #Lagrer posisjonen til vandrer nr. i i den i-ende raden i matrisa
    return alleVandrere                  #Returnerer matrisa
```

### Oppgave 1 e) - Tid før optimalisering

```python
tidfor = time.time()
flereVandrere(1000,1000, 0.5)
tidetter = time.time()
tidsendring = tidetter-tidfor
print(f'Tid før optimalisering: {tidsendring}')
```

### Oppgave 1 e) - Tid etter optimalisering

```python
# Ønsker å simulere 1000 virrevandrere med 1000 steg

def flereVandrereOpt(N, M, pR):  
    alleVandrere = np.zeros([M+1,N])
    for s in range(1,M+1):
        alleVandrere[s] = alleVandrere[s-1] + np.random.choice([-1,1],N,pR)
    return alleVandrere.T

tidforopt = time.time()
flereVandrereOpt(5,10, 0.5)
tidetteropt = time.time()
tidsendringopt = tidetteropt-tidforopt
print(f'Tid etter optimalisering: {tidsendringopt}')
print(f'Den optimaliserte koden går {tidsendring/tidsendringopt} ganger raskere enn den første')
```

Hvorfor simuleringen går raskere:
Jo færre operasjoner som må utføres, jo raskere vil koden kjøre. Dersom man greier å lager posisjoner direkte inn i en numpy array i stedet for å først opprette den og så fylle den, vil dette gjøre at koden kjører betraktelig raskere.

### Oppgave 1 f)

```python
M = 1000
N = 1000

oppgavef = flereVandrereOpt(N, M, 0.5).T
varians = np.zeros(M)
for i in range(M):
    varians[i]=statistics.variance(oppgavef[i]) # dette er variansen til posisjonen til virrevandrerne
    
    
def rettLinje(x, a): # a er den konstanten vi vil finne som tilpasser linja vår best
    return a*x

tid = np.arange(0,M,1)

popt, pcov = curve_fit(rettLinje, tid, varians) # pcov = kovarians som viser hvor bra grafen er tilpasset

plt.figure()
plt.plot(tid, varians, c = 'olive')
plt.plot(tid, rettLinje(tid, *popt), c = 'deeppink', label='fit: a=%5.3f' % tuple(popt)) 
#plotter en rett linje for å vise i hvor stor grad den andre minner om en rett linje

plt.xlabel('Tid', fontsize=15)
plt.ylabel('Varians', fontsize=15)
plt.title(f'Empirisk varians som funksjon av tid', fontsize=20)
plt.legend()
plt.show()
```

Vi observerer at variansen i stor grad ligner en rett linje. Variansen stiger når antall steg stiger. 

Stigningstallet til den tilpassede linja ligger rundt 1, noe som stemmer overens med svaret vi fikk i oppgave 1a).

Ser at når N er liten blir variansen større fordi det er større sanns. for at virrevandrerne kan bevege seg fra 
hverandre og dermed skape større varians. Når den er større vil variansen li mindre og mindre fordi det etter hvert 
vil være tilnærmet like mange som beveger seg til venstre som til høyre.
Når M er større vil det være enda flere steg, men fortsatt like mange virrevandrere, dermed vil det være større sjanse for 
at vi får de i samme posisjon slik at variansen er mindre, men vi må gjøre dette mange ganger, skik at det til
slutt ikke vil ha noe og si, og variansen vil altså være større. 


### Oppgave 1 g)

```python
from matplotlib import pyplot as plt
import numpy as np
def toDVandrere(N, M, pR, pU, dx, dy):                     # Funksjonen tar inn N vandrere, M tidssteg og sannsynlighet for å gå til høyre og opp
    dt = 1
    oppEllerBort = np.random.randint(0,2,M)                # Lager et array med tilfeldige antall 0 og 1. 0 tilsier horisontal forflytning, 1 tilsier vertikal
    posisjonx = np.zeros((N,M+1))                          # Lager en matrise med x-posisjoner med N rader(vandrere) og M+1 kolonner(tidspunkt) 
    posisjony = np.zeros((N,M+1))                          # Lager en matrise med y-posisjoner med N rader(vandrere) og M+1 kolonner(tidspunkt)
    tid = np.arange(0,M+1,dt)                              # Lager et array med tider, fra 0 til og med M
    
    
    for s in range(N):
        tilfeldigeTall = np.random.uniform(0,1,M)
        for t in range(1,M+1):
            if oppEllerBort[t-1] == 0:                     # Hvis vi får 0, beveger vi oss horisontalt
                if tilfeldigeTall[t-1]<pR:                 # Hvis tilfeldig sannsynlighet er mindre enn sannsynligheten for å bevege seg til høyre, går den til høyre   
                    posisjonx[s][t] = posisjonx[s][t-1]+dx[s][t-1]
                    posisjony[s][t] = posisjony[s][t-1]
                else:                                      # Hvis tilfeldig sannsynlighet er større enn sannsynligheten for å bevege seg til høyre, går den til venstre
                    posisjonx[s][t] = posisjonx[s][t-1]-dx[s][t-1]
                    posisjony[s][t] = posisjony[s][t-1]
            else:                                          # Hvis vi ikke får 0, beveger vi oss vertikalt
                if tilfeldigeTall[t-1]<pU:                 # Hvis tilfeldig sannsynlighet er mindre enn sannsynligheten for å bevege seg opp, går den opp
                    posisjony[s][t] = posisjony[s][t-1]+dy[s][t-1]
                    posisjonx[s][t] = posisjonx[s][t-1]
                else:                                      # Hvis tilfeldig sannsynlighet er større enn sannsynligheten for å bevege seg opp, går den ned
                    posisjony[s][t] = posisjony[s][t-1]-dy[s][t-1]
                    posisjonx[s][t] = posisjonx[s][t-1]
    
    return posisjonx, posisjony, tid


plt.figure()
antVandr = 10
varighet = 20
dx = np.ones((antVandr,varighet))
dy = dx
verdier = toDVandrere(antVandr,varighet,0.5,0.5, dx, dy)
plt.xlabel('Posisjon (x)', fontsize=15)
plt.ylabel('Posisjon (y)', fontsize=15)
plt.title(f'Virrevandring i 2D: x- og y-retning. pR=pU=0.5', fontsize=20)
for n in range(antVandr):
    plt.plot(verdier[0][n],verdier[1][n], c='deeppink',alpha=0.2)
plt.show()
```

### Oppgave 1 h)

```python
def krysserNull(vandrere, dim):
    if dim == 1:
        antallVandrere = (len(vandrere))
        tidssteg = (len(vandrere[0]))
        antall = np.zeros(len(vandrere[1]))
        
        for i in range(0,antallVandrere):
            sjekkVandrer = np.zeros(len(vandrere[1]))
            for t in range(1,tidssteg):
                if vandrere[i][t] == 0:
                    sjekkVandrer[t:] = 1
                    break
            antall = np.add(antall, sjekkVandrer)
        andel = antall/antallVandrere
        return andel
        
    elif dim == 2:
        antall = np.zeros(len(vandrere[2])) # dette er lengden av tiden returnert fra 2D funksjonen
        xPos = vandrere[0]
        yPos = vandrere[1]
        antallVandrere = len(xPos)
        tidssteg = len(xPos[0])

        
        for i in range(0, antallVandrere):
            sjekkVandrer = np.zeros(len(vandrere[2]))
            for t in range(2, tidssteg):
                if xPos[i][t] == 0:
                    if yPos[i][t] == 0:
                        sjekkVandrer[t:] = 1
                        break
            antall = np.add(antall, sjekkVandrer)
        andel = antall/antallVandrere
        return andel

krysserNull(toDVandrere(50,5000,0.5,0.5),2)
krysserNull(flereVandrereOpt(50,10,0.5),1)
```

```python
mm = 10**-3
org_dx = 0.004 #mm
original_dx = 0.004*3000
dt = 0.01 #sekunder

def D(dx,dt):               # Funksjon for utregning av diffusjonskonstanten. Tar inn dx i mm og dt i sekunder
    return (dx)**2/(2*dt)

print(f'Diffusjonskonstanten i friskt vev er: {D(0.004,0.01):.4f} mm^2*s^-1')
```

tumor k har red.koeff t_k < 1
reduerer delta x med sqrt(t_k)
lign for diffusjonskoeff. tilsier at da reduseres diff. med t_k

overlapping: reduksjon med sqrt(tk*tn...)

### Oppgave 2 b)

```python
def tumor(x, y,tumAreal,tumAntall,tumKoordXY,t_k,L): # Funksjonen tar inn x- og y-koordinater, arealet til tumorene, antall tumorer, koordinatene til sentrum og t_k
    tumRadius = np.sqrt(tumAreal/np.pi)
    deltaX = np.full((L,L),original_dx)
    for i in range(tumAntall):
        avstand = np.sqrt((tumKoordXY[i][0] - x)**2 + (tumKoordXY[i][1]- y)**2)
        deltaX = np.where(avstand<tumRadius, deltaX*np.sqrt(t_k),deltaX)
    return deltaX
    
L = 5
tumAntall = 10
tumKoordXY = np.random.randint(-N,N,(tumAntall,2))
tumAreal = 10
t_k = 0.1

#Lager x og y koordinater i meshgrid
nx, ny = (L, L)
x = np.linspace(-L, L, nx)
y = np.linspace(L, -L, ny)
xv, yv = np.meshgrid(x, y)



print(tumor(xv, yv,tumAreal,tumAntall,tumKoordXY,t_k,L))
```
