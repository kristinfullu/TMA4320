# Damptrykk for ${H_2O}$

## Innledning

En damptrykkurve er en fremstilling av forholdet mellom væske og gassfasen til et medium for ulike trykk og temperaturer. Dette prosjektet tar for seg damptrykkurven for vann. To viktige punkt på damptrykkurven er trippelpunktet og det kritiske punktet. Trippelpunktet merker ved hvilket trykk og hvilken temperatur vannets tre faser kan eksistere samtidig. Det kritiske punktet forteller oss at dersom trykket og temperaturen sammen stiger videre fra punktet, forsvinner forskjellen mellom væske og gass. For vann er trippelpunktet ved $273.1600K$, og det krisitke punktet ved $647.096K$

Vi skal også jobbe med trykk-volum diagram, ofte beskrevet som pV-diagram. Et 2D pV-diagram viser sammenhengen mellom trykk og volum ved en konstant temperatur, og har ulike inndelinger som beskriver hvilke ulike stadier av vannet som kan eksistere ved det trykket og volumet. Et pV-diagram skal blant annet konstrueres i oppgave 1 h).

Gjennom prosjektet skal vi benytte ulike numeriske metoder for å gjøre tilnærminger.

Den første numeriske metoden som brukes er Newtons metode, for både en og to ukjente. Gjennom en iterativ prosess finner metoden en stadig bedre tilnærmelse til et nullpunkt av den gitte funksjonen, noe som tilsier at funksjonen som skal løses er på formen $f(x) = 0$. Metoden kan ved hjelp av en Jacobi-matrise løse ligningssystemer, en egenskap som benyttes blant annet i oppgave 1 e).

Den andre metoden i bruk er Simpsons metode, som er en numerisk integrasjonsmetode. Metoden deler opp integrasjonsintervallet i flere delintervall, og approksimerer funksjonen i hvert delintervall.

Gjennom prosjektet skal vi gjøre ulike utregninger for verdier, blant annet for $p(T)$ og $V(T)$. For å forsikre oss om at disse verdiene er rett, sammenlignes de gjennom prosjektet ofte med eksperimentelle verdier fra EngineeringToolbox (1).

Målet med dette prosjektet er å bli kjent med ulike formler slik som van der Waals tilstandsligning og Clapeyrons ligning, og forstå hvorfor kjente metoder og ligninger av og til ikke er de beste tilnærmingene av faktiske forhold. Vi skal i tillegg se avvik som kan oppstå fra eksperimentelle verdier ved bruk av de ulike tilnærmingene.

### Oppgave 1 a)

Van der Waals tilstandsligning er gitt ved $$p = \frac{RT}{V-b}-\frac{a}{V^2}$$. For å kunne bruke denne er det viktig å bestemme parametrene a og b. Disse er konstante for et stoff, og kan regnes ut ved hjelp av bestemte verdier som tilhører det stoffet.

Fra side 4 i prosjektbeskrivelsen vet vi at $$ a = \frac{27R^2T_c^2}{64p_c}$$ og $$ b= \frac{RT_c}{8p_c}$$. Vi har også fått oppgitt de eksperimentelle verdiene $T_c = 647.096 K$ , $p_c = 22.064 MPa$ og $V_c = 55.948 mL$ for $H_2O$. I tillegg har det blitt opplyst at vi skal bruke at gasskonstanten har verdien $R = 8.314 \frac{J}{mol K}$.

Ved å sette inn verdiene i uttrykkene for a og b får vi
$$a = 0.5534 \frac{J^{2}}{mol^{2}Pa}$$

$$b = 3.048\cdot 10^{-5} \frac{J}{molPa}.$$

Dette er verdiene som vil brukes i videre utregninger med van der Waals ligning.

Når et stoff når den kristiske temperaturen $T_c$ og det kritiske trykket $p_c$, opptar ett mol av stoffet det som er definert som det kritiske volumet, $V_c$. For van der Waals tilstandsligning gjelder formelen $V_c = 3b$. 

Med vår utregnede verdi for b, får vi at $$V_c = 91.44 \frac{mL}{mol}.$$ Dersom vi sammenligner denne verdien med den eksperimentelle verdien for kritisk volum, 
$V_{c} = 55.948 \frac{mL}{mol}$, ser vi et klart avvik. Dette skjer fordi van der Waals tilstandsligning kun er en forenklet modell laget for å beskrive noen oppførsler hos ideell gass. Den tar ikke hensyn til mange av de komplekse egenskapene til ekte gasser, og svikter derfor ved utregningen av $V_c$.

Det er også verdt å bemerke at ettersom $V_c$ er definert som volumet ett mol stoff opptar, er det egentlig overflødig å ha med $mol$ i enheten $\frac{mL}{mol}$.

```python
# Samling av faste konstanter og variabler, og enkle omgjøringer
R = 8.314            #J/(mol K), gasskonstanten
a = 0.5534           #J^2/(mol^2*Pa)
b = 3.048*10**(-5)   #J/mol Pa
Tc = 647.096         #K, kritisk temperatur for vann
pc = 22.064*10**(6)  #Pa, kritisk trykk for vann

mL = 10**(-6)        #Ganger man med denne konverteres  mL til m^3, del for å gå andre veien
liter = 10**(-3)     #Ganger man med denne konverteres liter til m^3, del for å gå andre veien
bar = 10**(-5)       #Ganger man med denne konverteres pascal til bar, del for å gå andre veien
```

### Oppgave 1 b)


Van der Waals tilstandsligning brukes for å regne ut trykk, og dersom vi holder temperaturen konstant og varierer volumet, kan vi regne ut trykket som en funksjon av kun volumet. Dette er nyttig gjennom oppgave 1, og spesielt når vi skal konstruere det ferdige p-V fasediagrammet. Vi plotter derfor $p(V)$ vha. van der Waals tilstandsligning, og med $T = T_c$.

```python
def vdwCriticalTemp(minV,maxV,numVolumes):
    '''
    Input: Første og siste volumverdi, og antall volumverdier vi ønsker
    Output: Array med trykket ved hvert volum, og et array av volumene
    '''
    volumes = np.linspace(minV,maxV,numVolumes+1) 
    p = (R*Tc)/(volumes-b)-a/(volumes**2)
    return p, volumes

pTc, vTc = vdwCriticalTemp(75*mL,300*mL,225)  # Ganger for å gjøre om fra mL til m^3

fig = plt.subplots(1, figsize=(4, 4))
plt.plot(vTc/mL,pTc*bar) # Gjør om til mL og bar for å sammenligne med figur fra oppgavetekst
plt.title(f'p(V) ved temperatur Tc = {Tc}K')
plt.xlabel("Volum, mL", fontsize=12)
plt.ylabel("Trykk, bar", fontsize=12)
plt.grid()
plt.show()
```

```python
def ligning14(t): # setter uttrykket lik 0
    return (np.sinh((2)/t)**2) - 1

def ligning14Der(t): 
    return -(np.cosh(2/t)*4*np.sinh(2/t))/t**2

ligning15 = 2/(np.log(1+np.sqrt(2)))
```

```python
def newton(x0, f , fDer, maxIter, tol):
    
    # Lager en array
    x = np.zeros(maxIter)
    iArray = np.zeros(maxIter)
    
    # Startverdier
    x[0] = x0
    
    i = 0
    while ((np.abs(x[i] - x[i-1]) > tol) and (i < maxIter-1)):
        x[i+1] = x[i] - f(x[i])/fDer(x[i])
        i += 1   
        iArray[i] = i     
            
    return x[:i], iArray[:i] # for å ikke få med masse nuller på slutten

t,i = newton(1,ligning14,ligning14Der,600,6e-10)
print(t[-1])
print(ligning15)
```

### Oppgave 1 e)

```python
# Ligning 11

def f11(Vv,Vg,T):
    return R*T*(1/(Vg - b) - 1/(Vv - b)) - a*(1/Vg**2 - 1/Vv**2)

def f11DerG(Vv,Vg,T):
    return (2*a)/Vg**3 - (R*T)/(Vg - b)**2

def f11DerV(Vv,Vg,T):
    return (R*T)/(b-Vv)**2 - (2*a)/Vv**3
```

```python
# Ligning 12

def f12(Vv,Vg,T):
    ledd1 = ((R*T)/(Vg - Vv))*np.log((Vg - b)/(Vv - b)) 
    ledd2 = (R*T)/(Vg - b)
    ledd3 = a*(1/(Vg*Vv) - 1/(Vg**2))
    return ledd1 - ledd2 - ledd3

def f12DerG(Vv,Vg,T):
    ledd1 = (a*(Vg - 2*Vv))/(Vg**3 * Vv)
    ledd2 = (R*T*(b - 2*Vg + Vv))/((Vg - b)**2 * (Vg - Vv))
    ledd3 = (R*T*(np.log((b - Vg)/(b - Vv))))/(Vg - Vv)**2
    return ledd1 - ledd2 - ledd3

def f12DerV(Vv,Vg,T):
    ledd1 = a/(Vg * Vv**2)
    ledd2 = (R*T)/((b - Vv)*(Vg - Vv))
    ledd3 = (R*T*(np.log((b - Vg)/(b - Vv))))/(Vg - Vv)**2
    return ledd1 + ledd2 + ledd3
```

Brukte wolfram alpha for å løse uttrykkene for de deriverte både for (11) og (12).

```python
def jacobiMatrix(x, fDerV, fDerG, hDerV, hDerG,T):
    fdv, fdg = fDerV(x[0], x[1],T), fDerG(x[0], x[1],T)
    hdv, hdg = hDerV(x[0], x[1],T), hDerG(x[0], x[1],T)
    jacobiMatrise = np.array([[fdv, fdg], [hdv, hdg]])
    return jacobiMatrise

def functionSet(x,f,h,T):
    f11, f12 = f(x[0],x[1],T), h(x[0],x[1],T)
    funksjoner = np.array([f11,f12])
    return funksjoner


def newton2Variables(v0, g0, f, h , fDerV, fDerG, hDerV, hDerG, maxIter, tol, T):
    x = np.array([v0,g0])
    i = 0
    delta = 1
    while (np.linalg.norm(delta) > tol) and (i < maxIter-1):
        j = jacobiMatrix(x,fDerV,fDerG,hDerV,hDerG,T)
        funk = functionSet(x,f,h,T)
        delta = np.linalg.solve(j, -funk)
        x += delta
        i += 1
    return x
```

```python
T = np.linspace(274,647,374)

vvArray = np.zeros(len(T))
vgArray = np.zeros(len(T))

vvGuess = 35.7*mL    #m^3
vgGuess = 12.6*liter #m^3
for i in range(len(T)):
    vvArray[i],vgArray[i] = newton2Variables(vvGuess,vgGuess,f11,f12,f11DerV,f11DerG,f12DerV,f12DerG,600,6e-10,T[i]) # fyller inn verdiene for vg og vv for ulike T
    vvGuess = vvArray[i] # må legge til et nytt gjett for neste temperatur, slik at det skal passe hvordan det utvikler seg
    vgGuess = vgArray[i]

```

```python
# def polynom4_log(x, ag, bg, cg, dg, eg):
#     return ag*x**4 + bg*x**3 + cg*x**2 + dg*x +eg

def polynom4(x, ag, bg, cg, dg, eg):
    return ag*x**4 + bg*x**3 + cg*x**2 + dg*x +eg

def polynom2(x, av, bv, cv):
    return av*x**2 + bv*x + cv

def proporsjonal(x, al, bl):
    return al*np.abs(x-Tc)**bl

popt, pcov = curve_fit(polynom4, kelvinG, np.log(Vg)) #Finner for log for det er enklere og se feilen
popt2, pcov2 = curve_fit(polynom2, kelvinL, np.log(Vv)) #Finner for log for det er enklere og se feilen
popt3, pcov3 = curve_fit(proporsjonal, kelvinH, L)

```

### Opggave 2c)

```python
def VgFunk(x):
    return np.exp(popt[0]*x**4 + popt[1]*x**3 + popt[2]*x**2 + popt[3]*x + popt[4])

def VvFunk(x):
    return np.exp(popt2[0]*x**2 + popt2[1]*x + popt2[2])

def LFunk(x):
    return popt3[0]*np.abs(x-Tc)**popt3[1]
```

```python

def pDeriv(x):
    return LFunk(x)/(x*(VgFunk(x) - VvFunk(x))) 

def simpson(f, Tstart, Tend, n, m):
    """Tarf: funksjonen som skal integreres
    Tstart: startpunktet
    Tslutt: sluttpunkt
    n: antall steg for hver simpsonsfunksjoninterering
    m: antall ganger simpsonsfunksjon skal kjøre """
    x = np.linspace(Tstart, Tend, m)
    h = (x[1] - x[0])/n
    s = np.zeros(m)
    for i in range(1, m):
        x1 = np.linspace(x[i-1], x[i], n)
        s1 = f(x[i]) + f(x[i-1])
        for j in range(1, n-1):
            if j % 2 == 0:
                s1 += 2 * f(x1[j])
            else:
                s1 += 4 * f(x1[j])
        s[i-1] = s1
    
    return h*s/3, x

p, temps= simpson(pDeriv, 274, 647, 10, 374)

```

```python
# Integrerer ligning (13) ved bruk av interpolasjonene og 2c)
def pDerivCubic(x):
    return LCubic(x)/(x*(VgCubic(x) - VvCubic(x)))

pInterpolasjon, temps = simpson(pDerivCubic, 274, 647, 10, 374)

# Plot
plt.plot(T, pInterpolasjon, label="Interpolasjon")
plt.plot(exptemps, exppressures*1000, 'o', label="Eksperimentell")
plt.title("Koeksistenskuren $p(T)$ med den eksperimentelle måleserien")
plt.xlabel("$T, K$")
plt.ylabel("Damptrykk, $Pa$")
plt.legend()
plt.show()
```

Plott koeeksistenskurven p(T) for alle metodene dere har brukt (van der Waals og

i ligning (6) med konstant L (finn en verdi L for som tilnærmer den eksperimentelle
m˚aleserien godt og bestem selv referansepunktet (p0, T0)). Inkluder ogs˚a den eksperimentelle m˚aleserien.

```python
# Referansepunkter, det er ved disse verdiene det skifter fra væske til gass
p0 = 100*1000
t0 = 374
l = 40650
T = np.linspace(274, 647, 374)
R = 8.3145

# Alle metodene vi ønsker å sammenligne
pVvVdW = vdwNew(VvExp,TVolum)                       # Van der Waals
pVgVdW = vdwNew(VgExp,TVolum)                       # Van der Waals
pSimp, temp = simpson(pDeriv, t0, T[-1], 10, 374)   # Simpsons metode 

def pIdeell(x, l, p0, t0):
    pAnalytisk = np.zeros(len(x)) 
    for i in range(len(x)):
        pAnalytisk[i] = p0 * np.exp(l/(R*t0) - l/(R*x[i]))
    return pAnalytisk
pIdeellPlot = pIdeell(T, l, p0, t0)                 # Ideell gass

# Plot for alle metodene
plt.plot(T, pVv, label="$p_{V_v}$ van der Waals") 
plt.plot(T, pVg, label="$p_{V_g}$ van der Waals") 
plt.plot(T, pSimp, label="Simpsons")
plt.plot(T, pInterpolasjon, label="Simpsons cubic spline")
plt.plot(T, pIdeellPlot, label="Ideell gass")
plt.plot(exptemps, exppressures*1000, 'o', label="Eksperimentell")
plt.title("Sammenligning av metodene")
plt.xlabel("Temperatur, $K$")
plt.ylabel("Trykk, $Pa$")
plt.legend()
plt.show()
```
